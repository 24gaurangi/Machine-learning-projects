```python
# Library Imports

import scipy.io 
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from collections import defaultdict

# Extracting Dataset

data=scipy.io.loadmat('Project_2_AllSamples.mat')['AllSamples']
obj_func_combined=[]

# Functions used in the code

# 1.Function to assign classes to data points based on given cluster centroids
# Inputs: Dataset with data points(D), list of cluster centroids(CE)
# Output: List of classes assigned to data points (classify)

def assign(D, CE):
    classify=[]
    for i in range(len(D)):
        min=1000000000
        minindex=0;
        for j in range(len(CE)):
            distance=math.pow(CE[j][0]-D[i][0],2)+math.pow(CE[j][1]-D[i][1],2)
            if distance<min:
                min=distance
                minindex=j
        classify.append(minindex)
    return classify

# 2.Function to define cluster centroids from given data points and their classes
# Inputs: Dataset with data points(D), list of cluster classes assigned to data points(C)
# Output: List of new cluster centroids

def centroid(D, C):
    dict = defaultdict(list)
    for i in range(len(C)):
        dict.setdefault(C[i],dict[C[i]].append(D[i]))
    for i in dict:
        dict[i]=[sum([x[0] for x in dict[i]])/float(len(dict[i])),sum([x[1] for x in dict[i]])/float(len(dict[i]))]
    return [dict[i] for i in dict]

# 3.Function to compare if two lists are equal
# Inputs: Two lists (a and b)
# Output: True if lists are identical, False if not

def list_compare(a,b):
    a.sort()
    b.sort()
    for i in range(len(a)):
        if a[i]!=b[i]:
            return False
    return True

# 4.Function to calculate Euclidean distance between two data points
# Inputs: Data points(x and y)
# Output: Euclidean distance between a and y

def euclid_dist(x,y):
    return math.sqrt(math.pow(x[0]-y[0],2) + math.pow(x[1]-y[1],2))


# 5.Function to calculate Objective function value for given inputs
# Inputs: Dataset with data points(D), list of cluster classes assigned to data points(C) and list of cluster centroids(CE)
# Output: Objective function value

def obj(D, C, CE):
    dict = defaultdict(list)
    for i in range(len(C)):
        dict.setdefault(C[i],dict[C[i]].append(D[i]))
    for i in dict.keys():
        dict[i]=sum([math.pow(euclid_dist(x,CE[i]),2) for x in dict[i]])
    return sum([dict[x] for x in dict.keys()])


# 6.Function to calculate farthest data point from input data to the centroids 
# Inputs: Dataset with data points(D), list of cluster centroids(CE)
# Output: Index of the farthest point(far_index)

def farthest(D,CE):
    data_dict=defaultdict(list)
    far_index=0
    for i in range(len(D)):
        for j in range(len(CE)):
            data_dict.setdefault(i,data_dict[i].append(euclid_dist(D[i], CE[j])))
        data_dict[i]=sum(data_dict[i])
    for l in data_dict.keys():
        if data_dict[l]==max(data_dict.values()):
            far_index=l
    return far_index
```

```python
# Strategy 2: pick the first center randomly; for the i-th center (i>1), choose a sample
# (among all possible samples) such that the average distance of this chosen one to all
# previous (i-1) centers is maximal.

# Running 2 iterations for the strategy
for iter in range(2):
    # Variables : 
    # N => length of input dataset
    # X, Y => Input data points split into arrays of X-co-ordinates and Y-co-ordinates for plotting
    # objfunction => Dictionary to store objective function value for each cluster size from 2 to 11.

    N=len(data)
    X, Y=list(zip(*data))
    objfunction_strategy2={}

    # Looping for number of clusters from 2 to 11

    for k in range(2,11):
        print('Total number of clusters - ',k)
        print('*****************************************************************************************')

        # Assign first center randomly
        index =random.randint(0,N-1)
        centroids=[data[index].tolist()]

        # For next k-1 centroids, assign a farthest sample
        for c in range(k-1):
            m=farthest(data.tolist(),centroids)
            centroids.append(data.tolist()[m])


        # Classify input data points based on the randomly picked centroids
        categories=assign(data, centroids)

        # Calculate new centroids for the new cluster classes
        new_centroids=centroid(data.tolist(), categories)

        # Repeat the above 2 steps until the centroids do not change any further
        count=0
        while not list_compare(centroids,new_centroids):
            categories=assign(data, new_centroids)
            centroids=new_centroids
            new_centroids=centroid(data.tolist(), categories)
            count+=1


        # Plot the final centroids
        xn, yn=zip(*new_centroids)
        plt.scatter(X, Y, c=categories)
        plt.scatter(xn,yn, color='red')
        plt.show()

        # Calculate Objective function value for the cluster size k
        obj_value= obj(data.tolist(),categories, new_centroids)
        print('*****************************************************************************************')
        print('Total number of iterations - ', count)    
        print('Final centroids :\n',new_centroids)
        print('Objective function: ',obj_value)
        print('*****************************************************************************************')
        objfunction_strategy2.setdefault(k,obj_value)

    print('All obj function values for respective cluster sizes are as below -\n',objfunction)
    obj_func_combined.append(objfunction_strategy2)
```
